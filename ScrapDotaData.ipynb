```
def getHeroIds(listaDict):
  lstTeam0 = []
  lstTeam1 = []
  for i in listaDict:
    if i['is_pick']:
      if i['team'] == 0:
        lstTeam0.append(i['hero_id'])
      elif i['team'] == 1:
        lstTeam1.append(i['hero_id'])
  return lstTeam0, lstTeam1
```

```
lstWin = []
lstTeam0 = []
lstTeam1 = []
count = 0
for i in lstDataMatch:
  print(count)
  count += 1
  if i['pick_bans'] is None:
    continue
  lstWin.append(i['radiant_win'])
  team0, team1 = getHeroIds(i['pick_bans'])
  lstTeam0.append(team0)
  lstTeam1.append(team1)
```

```
dfDraft = pd.DataFrame({'radiant_win' : lstWin, 'team_radiant' : lstTeam0, 'team_dire' : lstTeam1})
```

```
def getHeroNames(lstIds):
  lstHeroes = []
  print(lstIds)
  for i in lstIds:
    hero = dfHero[dfHero['id'] == i]['name'].values[0]
    lstHeroes.append(hero)
  return lstHeroes
```

```
dfDraft['heroes_radiant'] = dfDraft.team_radiant.apply(lambda x: getHeroNames(x))
```

```
dfDraft['heroes_dire'] = dfDraft.team_dire.apply(lambda x: getHeroNames(x))
```
